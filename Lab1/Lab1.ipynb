Przedstawić sposoby łączenia ramek danych za pomocą metod merge i concat

```python
dane1 = pd.read_csv('Dane1.csv', encoding='latin1')
dane1
```

```python
pd.merge (dane, dane1, on = ['iso3'], how = 'inner') # bierze tylko wiersze, które pasują do obu ramek danych
```

```python
pd.merge (dane, dane1, on = ['iso3'], how = 'outer') # wszystkie wiersze ze wszystkich ramek danych, nie ma znaczenia, czy pasują, czy nie
```

```python
pd.merge (dane, dane1, on = ['iso3'], how = 'left') # tylko pasujące wiersze + wszystkie wiersze z lewej ramki danych
```

```python
pd.merge (dane, dane1, on = ['iso3'], how = 'right') #tylko pasujące wiersze + wszystkie wiersze z prawej ramki danych
```

Przedstawić możliwości pracy z dużymi plikami przy użyciu argumentu chunksize

```python
for chunk_dane in pd.read_csv('dane.csv',encoding='latin1',chunksize = 50000):
    print("CHUNK DF")
    print(chunk_dane.head())
```

```python
new_dane = pd.DataFrame() # pusta ramka danych
for chunk_dane in pd.read_csv('dane.csv',encoding='latin1',chunksize = 50000):
    result = chunk_dane.groupby(['iso3', 'year']).agg({'the_total_lower': 'mean',
                                                    'the_total_upper': 'mean'})
    new_dane = pd.concat([new_dane,result])

new_dane
```
